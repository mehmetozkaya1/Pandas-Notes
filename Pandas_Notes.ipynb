## Creating Pandas Series:

```python
mydict = {"Mehmet":40,"Aybala":50,"Güldane":60}
pandaS1 = pd.Series(mydict)
pandaS1
```

We can create pandas series by the help of the 'Series' command. We can convert dictionaries into Pandas Series.

```python
myAges = [20,30,40]
myNames = ["Mervan","Emin","Mehmet"]
pandaS2 = pd.Series(myAges, myNames)
pandaS2
```

We can also convert two array into a Pandas Series.

```python
pandaS3 = pd.Series(data=myAges, index=myNames)
pandaS3
```

We can also adjust the indexes and the datas of the table by the help of the parameters 'data' and 'index'

```python
numpy_array = np.arange(5)
pandaS4 = pd.Series(numpy_array)
type(pandaS4)
```

Numpy arrays can be converted into Pandas Series.

```python
pandaS5 = pd.Series([80,90,100],["Student1","Student2","Student3"])
pandaS5
```

Pandas Series can be created with just giving the values directly.

#### Working with missing values:

Let's create a dataframe with missing values.

```python
dictionary = {"Çorum": [30,29,np.nan],"Aksaray": [20,np.nan,25],"Yozgat": [40,39,38]}

last_df = pd.DataFrame(dictionary)
last_df
```

```python
last_df.isnull()
```

'isnull' method returns a boolean value for each of the values in the dataframe. If the values is NaN, it returns True; if not, it returns False.

```python
last_df.dropna()
```

'dropna' method drops all the rows with NaN values. It doesn't affect the original dataframe unless you use inplace parametre. 'dropna''s default axis values is 0 so it deletes row. If you want to delete columns you should adjust the axis to 1.

```python
last_df.dropna(axis=1)
```

```python
new_values = {"İstanbul": [30,29,np.nan],"Ankara": [20,np.nan,25],"İzmir": [40,39,38],"Antalya": [45,np.nan,np.nan]}
new_df1 = pd.DataFrame(new_values)
new_df1
```

```python
new_df1.dropna(axis=1,thresh=2)
```

It only drops rows if there are 2 or more than 2 NaN values. (thresh parameter)

```python
new_df1.fillna(30)
```

'fillna' methos assigns the given value to all NaN values.

#### Groupby method:

```python
salaries = {"Departmant" : ["Software","Software","Cleaning","Cleaning","Law","Law"],
              "Worker" : ["Ahmet","Mehmet","Osman","Burak","Zeynep","Gülsüm"],
               "Salary" : [100,150,200,300,400,500]
              }

df = pd.DataFrame(salaries)
df
```

```python
group = df.groupby("Departmant")
group
```

We created an object grouped by 'department'.

```python
group.count()
```

It allows us to see how many values ​​are in the object according to the department.

```python
group.mean()
```

Averages each department for numeric values. (For this example, it averaged salaries.)

```python
group.sum()
```

Sums each department for numeric values. (For this example, it sums salaries.)

```python
group.min()
```

Finds the minimum value for each department for numeric and alphabetic values.

```python
group.max()
```

Finds the maximum value for each department for numeric and alphabetic values.

```python
group.describe()
```

It calculates all the values ​​that we can reach with the 'describe' method for each department.
